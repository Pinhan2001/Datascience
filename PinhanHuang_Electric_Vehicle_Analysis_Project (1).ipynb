## Data Cleaning
Before conducting analysis, it's important to clean the data. This includes:
- Checking for missing values.
- Handling inconsistent data entries.
- Dropping irrelevant columns if necessary.
---


```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'Electric_Vehicle_Population_Data.csv'
df = pd.read_csv(file_path)

# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)

# Drop rows with missing data in critical columns
df = df.dropna(subset=['Electric Vehicle Type', 'Model Year', 'Make'])

# Fill missing values in non-critical columns if necessary (e.g., location data)
df['County'] = df['County'].fillna('Unknown')

# Ensure all Model Year values are integers
df['Model Year'] = df['Model Year'].astype(int)

# Verify the cleaned data
print("\nData after cleaning:")
print(df.info())
```

## Exploratory Data Analysis (EDA)

### 1. Trends in Electric Vehicle Types


```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
# Analyze trends in EV types
ev_type_counts = df['Electric Vehicle Type'].value_counts()

# Plot the distribution of electric vehicle types
plt.figure(figsize=(8, 6))
ev_type_counts.plot(kind='bar', color=['skyblue', 'lightgreen'], edgecolor='black')
plt.title('Distribution of Electric Vehicle Types', fontsize=16)
plt.xlabel('Electric Vehicle Type', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

### 2. Trends Over Time


```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
# Group data by model year and count the number of vehicles
yearly_trends = df.groupby('Model Year')['Electric Vehicle Type'].value_counts().unstack()

# Plot the trends over time
plt.figure(figsize=(10, 6))
yearly_trends.plot(kind='line', marker='o', ax=plt.gca())
plt.title('EV Trends Over Time by Type', fontsize=16)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)
plt.legend(title='EV Type', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 3. Geographical Analysis


```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
# Count vehicles by county
county_counts = df['County'].value_counts().head(10)  # Top 10 counties

# Plot vehicle distribution by county
plt.figure(figsize=(10, 6))
county_counts.plot(kind='bar', color='coral', edgecolor='black')
plt.title('Top 10 Counties by EV Registration', fontsize=16)
plt.xlabel('County', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 4. Manufacturer Dominance

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
# Count vehicles by manufacturer
manufacturer_counts = df['Make'].value_counts().head(10)  # Top 10 manufacturers

# Plot manufacturer dominance
plt.figure(figsize=(10, 6))
manufacturer_counts.plot(kind='bar', color='goldenrod', edgecolor='black')
plt.title('Top 10 Manufacturers by EV Registration', fontsize=16)
plt.xlabel('Manufacturer', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
